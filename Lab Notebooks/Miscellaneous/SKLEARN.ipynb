```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv('placement.csv')
df_binary=df[['cgpa','placement_exam_marks']]
df_binary.columns=['cgpa','marks']
df_binary.head()

sns.lmplot(x='cgpa',y='marks',data=df_binary)
plt.show

#for NAN or missing values
#df_binary.fillna(method ='ffill', inplace = True)

X=np.array(df_binary['cgpa']).reshape(-1,1)
y=np.array(df_binary['marks']).reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

op = LinearRegression()
op.fit(X_train, y_train)
acc = op.score(X_test, y_test)
print('Accuracy of model:',acc)
print('Coefficient:', op.coef_)
print('Intercept:', op.intercept_)

y_pred = op.predict(X_test)
plt.scatter(X_test, y_test, color ='y')
plt.plot(X_test, y_pred, color ='k')
plt.show()

```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df = pd.read_csv('Salary_dataset.csv')
df_binary=df[['YearsExperience','Salary']]
df_binary.columns=['exp','salary']
df_binary.head()

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

sns.regplot(x='exp',y='salary',data=df_binary, ax=axes[0])
axes[0].set_title('Dataset Plot')

#for NAN or missing values
#df_binary.fillna(method ='ffill', inplace = True)

X=np.array(df_binary['exp']).reshape(-1,1)
y=np.array(df_binary['salary']).reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

op = LinearRegression()
op.fit(X_train, y_train)
acc = op.score(X_test, y_test)
print('Accuracy of model:',acc)
print('Coefficient:', op.coef_)
print('Intercept:', op.intercept_)

y_pred = op.predict(X_test)
axes[1].scatter(X_test, y_test, color ='y')
plt.plot(X_test, y_pred, color ='b')
axes[1].set_title('Regression Plot')
plt.show()


```
