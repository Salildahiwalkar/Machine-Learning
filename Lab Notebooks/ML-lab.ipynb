# **Numpy**

## Arrays


```python
import numpy as np
arr = np.arange(10000)
print(arr)
print(arr[:10])
print(len(arr))
```

## Array vs List

```python
arr = np.arange(10000)
arr_list = list(range(10000))
%time a = [x * 10 for x in arr_list]
%time b = arr * 10
```

It takes less time to multiply an array with a number than list.

## Multidimensional Arrays

```python
arr = np.random.randn(4, 4)
print(arr)
arr3 = arr * 3
print(arr+arr3)
```

Choosing specific element
 - choosing element from first row third column

```python
arr = np.random.randn(4, 4)
print(arr)
arr[0, 2]
```

Slicing
 - first two rows


```python
arr = np.random.randn(4, 4)
print(arr)
arr[0:2]
```

 - middle section

```python
arr = np.random.randn(4, 4)
print(arr)
arr[0:2, 1:3]
```

# **Pandas**

Pandas defines two majore data structures, Series and DataFrame. A Series is a one-dimensional array-like object containing a sequence of values.

```python
import pandas as pd
pd.Series(['a', 'b', 'c', 'd', 'e'])
```

Series can also have indexes or labels attached to each value (where they are then almost similar to a dict)

```python
ser = pd.Series(['a', 'b', 'c', 'd', 'e'], index=['foo', 'bar', 'more', 'note', 'extra'])
ser
```

Now you can access each object using the dot notation or brackets
 - ser.foo
 - ser['foo']

```python
import pandas as pd
ser = pd.Series(['a', 'b', 'c', 'd', 'e'], index=['foo', 'bar', 'more', 'note', 'extra'])
ser.to_dict()
```
