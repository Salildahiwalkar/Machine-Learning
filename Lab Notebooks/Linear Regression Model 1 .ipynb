```python
import numpy as np
import matplotlib.pyplot as plt

x_train = np.array([1.0, 2.0, 3.0, 5.0, 8.0])
y_train = np.array([200.0, 500.0, 800.0, 900.0, 1100.0])

#prediction
def compute_model(x,w,b):
  n=x.shape[0]
  func = np.zeros(n)
  for i in range(n):
    func[i]=w*x[i]+b
  return func

#cost function
def cost_function(x,y,w,b):
    n=x.shape[0]
    cost = 0
    for i in range(n):
      j_wb=w*x[i]+b
      diff = j_wb-y[i]
      cost = cost + (diff)**2
      total_cost=1/(2*n)*cost
      print(diff, cost, total_cost)
    return total_cost

def model(x,y):
  n=x.shape[0]
  min_cost = float('inf')
  best_w = None
  best_b = None

  for i in range(0,1000):
    w=i
    for j in range(0,500):
      b=j
      cost=cost_function(x,y,w,b)
      if cost < min_cost:
        min_cost = cost
        best_w = w
        best_b = b
  return best_w, best_b, min_cost

w, b, cf = model(x_train,y_train)
print("Best W,B for minimum Cost Function : ",w, b, cf)
```
